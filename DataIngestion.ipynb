````python
from codebleu import calc_codebleu

prediction = test_set.iloc[0]["output"]
reference = test_set.iloc[1]["output"]

# We use CodeBLEU with dataflow_match_score set to 0 and on the javascript language (hoping that generalizes to typescript)
result = calc_codebleu([prediction], [reference], lang="python", weights=(0.25, 0.25, 0.25, 0.25), tokenizer=None)
print(result)
# {
#   'codebleu': 0.5537, -
#   'ngram_match_score': 0.1041,
#   'weighted_ngram_match_score': 0.1109,
#   'syntax_match_score': 1.0,
#   'dataflow_match_score': 1.0
# }
````

````python
import matplotlib.pyplot as plt
import numpy as np

# Data for the models
models = [
    '7B (Base)',
    '13B (Base)',
    '7B (Fine-Tuned)',
    '13B (Fine-Tuned)'
]

# Scores for CodeBLEU Avg and CodeBLEU Max
codebleu_avg = [0.147, 0.140, 0.161,  0.178]
codebleu_max = [0.508, 0.673, 0.692, 0.624]

# Setting the positions and width for the bars
pos = np.arange(len(models))
bar_width = 0.35

# Plotting CodeBLEU Avg
plt.bar(pos, codebleu_avg, bar_width, label='CodeBLEU Avg', alpha=0.5)

# Plotting CodeBLEU Max
plt.bar(pos + bar_width, codebleu_max, bar_width, label='CodeBLEU Max', alpha=0.5)

# Adding labels and title
plt.xlabel('Models')
plt.ylabel('Scores')
plt.title('Comparison of CodeBLEU Scores')

# Adding legend
plt.legend()

# Setting the position of the x ticks
plt.xticks(pos + bar_width / 2, models, rotation=90)

# Setting the y-axis limits
plt.ylim(0, max(codebleu_max) + 0.1)

# Display the plot
plt.tight_layout()
plt.show()

# Updated model names
models_short = [
    '7B (Base)',
    '7B (Fine-Tuned)',
    '13B (Base)',
    '13B (Fine-Tuned)'
]

# Setting the positions and width for the bars
pos = np.arange(len(models_short))

# Creating the bar chart again with updated names and title
plt.figure(figsize=(10, 6))  # Increase figure size for better readability
plt.bar(pos, codebleu_avg, bar_width, label='CodeBLEU Avg', alpha=0.5)
plt.bar(pos + bar_width, codebleu_max, bar_width, label='CodeBLEU Max', alpha=0.5)

plt.xlabel('Models')
plt.ylabel('Scores')
plt.title('Comparison of CodeBLEU Scores on CodeLlama-Instruct Models')

plt.legend()
plt.xticks(pos + bar_width / 2, models_short)  # Rotate the model names to be horizontal
plt.ylim(0, max(codebleu_max) + 0.1)

plt.tight_layout()
plt.show()

````

````python
# HumanEval scores for the models
humaneval_scores = [34.8, 42.7, 34.3, 40.2]

# Creating the bar chart for HumanEval scores
plt.figure(figsize=(10, 6))  # Increase figure size for better readability
plt.bar(models_short, humaneval_scores, color='skyblue')

plt.xlabel('Models')
plt.ylabel('HumanEval Scores')
plt.title('Comparison of HumanEval Scores on CodeLlama-Instruct Models')

# Rotate the model names to be horizontal (already horizontal by default)
plt.xticks(models_short)

plt.ylim(0, max(humaneval_scores) + 5)  # Adding a little space above the highest bar

plt.tight_layout()
plt.show()
````
